```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
data=pd.read_csv("泰坦尼克/train.csv")
# for i in range(1,data.shape[1]):
#     value=data.iloc[:,i].values
#     for i in value:
#         if 
value=data.iloc[:,5:6].values
npvalue=np.array(value)
vlist=[]
nanindex=[]
for i in npvalue:
    vlist.append(i[0])
for i in range(len(vlist)):
    if math.isnan(vlist[i])==True:
        nanindex.append(i)
for i in nanindex:
    del vlist[i]
    npvalue1=np.array(vlist)
    npvalue1_med=np.median(npvalue1)
    npvalue[i]=npvalue1_med


    
print(npvalue)
```
