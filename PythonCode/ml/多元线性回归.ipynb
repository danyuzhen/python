```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
#--------------------导入数据---------------------
data=pd.read_csv("MachineLearning\\Part 2 - Regression\\Section 5 - Multiple Linear Regression\\50_Startups.csv")
#-------------------
#除去最后一列的值
x=data.iloc[:,0:4].values
#最后一列标签
y=data.iloc[:,-1].values
# print(x)
# print(y)
# #--------------------缺失数据（可选）---------------------
# imputer=Imputer(missing_values="NaN",strategy="mean",axis=0)
# imputer=imputer.fit(x[:,1:3])
# x[:,1:3]=imputer.transform(x[:,1:3])
# #--------------------分类数据（可选,包含虚拟编码）---------------------
labelcode_x=LabelEncoder()
labelcode_y=LabelEncoder()
x[:,3]=labelcode_x.fit_transform(x[:,3])
oneoht=OneHotEncoder(categorical_features=[3])
x=oneoht.fit_transform(x).toarray()
x=x[:,1:]
#############因变量处理######################
# y=labelcode_y.fit_transform(y)
# print(x)
# print(y)
#--------------------拆分训练集数据集---------------------
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
#--------------------线性回归拟合all in---------------------
regressor=LinearRegression()
regressor.fit(xtrain,ytrain)
ypred=regressor.predict(xtest)
#-------------------反向淘汰--------------------------------
xtrain=np.append(arr=np.ones((40,1)),values=xtrain,axis=1)
xopt=xtrain[:,[0,1,2,3,4,5]]
regressorOLS=sm.OLS(endog=ytrain,exog=xopt).fit()
regressorOLS.summary()
xopt=xtrain[:,[4,5]]
regressorOLS=sm.OLS(endog=ytrain,exog=xopt).fit()
regressorOLS.summary()
```
